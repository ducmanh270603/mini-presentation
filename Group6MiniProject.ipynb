## Analysis


### 1. About the workforce in 2021-2022

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
labels = ["2021", "2022"]
city_population = [1853500, 1921160]
countryside_population = [3202550, 3249329]
nbars = len(labels)
y = np.arange(nbars)
width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(y - width/2, city_population, width, label='City')
rects2 = ax.bar(y + width/2, countryside_population, width, label='Countryside')
ax.set_xlabel("Year")
ax.set_ylabel("Population")
ax.set_title("Workforce")
ax.set_xticks(y)
ax.set_xticklabels(labels)
ax.legend()
max_population = max(max(city_population), max(countryside_population))
ax.set_ylim([0, max_population + max_population*0.1])
for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate(f"{height}", xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points",
                ha='center', va='bottom')
plt.ylim(0, 4500000)
plt.show()
```

### Unemployment rate in working age

```python
labels = ['City', 'Countryside']
percent_2021 = [4.33, 2.50]
percent_2022 = [2.82, 2.04]
x = range(len(labels))
plt.bar(x, percent_2021, width=0.4, label='2021', color='lightblue')
plt.bar(x, percent_2022, width=0.4, label='Preliminary 2022', color='orange')
plt.xlabel('Local Type')
plt.ylabel('Percent (%)')
plt.title('Unemployment percentage change from 2021 to 2022 preliminary')
[plt.text(i, percent_2021[i] + 0.1, f'{percent_2021[i]}%', ha='center', color='black') for i in x]
[plt.text(i, percent_2022[i] + 0.1, f'{percent_2022[i]}%', ha='center', color='black') for i in x]
plt.grid(True, linestyle='--', alpha=0.5)
plt.ylim(0, 5)
plt.legend()
plt.xticks(x, labels)
plt.show()
```

### 2. Unemployment rate divided by age

```python
x=["2018","2019","2020","2021","Preliminary 2022"]
yall=[2.19,2.17,2.48,3.20,2.34]
y15=[6.92,6.51,7.21,8.55,7.78]
y25=[1.49,1.59,1.99,2.66,1.77]
y50=[0.99,0.70,0.99,1.86,1.17]

plt.plot(x,yall,'-r',label='Totals')
plt.plot(x,y15,':o',label='From 15-24 years old')
plt.plot(x,y25,'--g',label='From 25-49 years old')
plt.plot(x,y50,'-p',label='From 50 years old or older')

plt.xlabel('Year')
plt.ylabel('Age')
plt.title('Unemployment rate in the working age group by age group divided by Age Group and Year')
plt.legend()
plt.show()
```

### 3. Unemployment rate divided by degree level

```python
trinh_do = ["No degree", "High school", "College", "University", "Master", "PhD"]
ty_le_2021 = [3.20,3.25,2.00,2.66,4.43,3.38]  # Dữ liệu năm 2021 
ty_le_2022 = [2.34,2.21,1.60,2.31,3.41,3.16]  # Sơ bộ năm 2022

x = np.arange(len(trinh_do))

width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, ty_le_2021, width, label='2021', color='skyblue')
plt.bar(x + width/2, ty_le_2022, width, label='2022', color='lightcoral')

plt.xlabel('Academic level')
plt.ylabel('Unemployment rate (%)')
plt.title('Unemployment rate by degree level for 2020 and 2021')
plt.xticks(x, trinh_do, rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

### 4. Unemployment rate divided by men and women

```python
years = [2021, 2022]
male_unemployment_rates = [3.15, 2.36]
female_unemployment_rates = [3.26, 2.32]

bar_width = 0.4
index = np.arange(len(years))

plt.bar(index, male_unemployment_rates, bar_width, label='Male')
plt.bar(index + bar_width, female_unemployment_rates, bar_width, label='Female')

plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate of Males and Females in 2021 and 2022')
plt.xticks(index + bar_width/2, years)
plt.legend()
plt.show()
```
